# Training ACT with LeRobot
이 명령어는 lerobot 라이브러리의 lerobot_train.py 스크립트를 실행하여 robot control policy을 학습시킵니다.

model policy에 대한 내용은 아래 링크에 있습니다.

https://huggingface.co/docs/lerobot/act


다음 명령어들을 본인의 설정에 맞게 조정하시면 됩니다:

1. `--dataset.repo_id=YOUR_HF_USERNAME/YOUR_DATASET`: 데이터셋이 저장된 Hugging Face Hub 저장소 ID로 변경하세요. 예: taeyoungss/Model_dataset3

2. `--policy.type=act`: 사용할 모델 정책 구성을 지정합니다.

3. `--output_dir=outputs/train/...`: 학습 로그와 모델 체크포인트가 저장될 디렉토리입니다.

4. `--job_name=...`: 이 학습 작업의 이름으로, 로깅 및 Weights & Biases에서 사용됩니다.

5. `--policy.device=cuda`: NVIDIA GPU에서 학습하는 경우 `cuda`를 사용하세요. Apple Silicon의 경우 `mps`, GPU가 없는 경우 `cpu`를 사용하세요.
 colab에서 학습 할 경우 cuda를 그대로 설정 하시면 됩니다

6. `--wandb.enable=true`: 학습 진행 상황 시각화를 위해 Weights & Biases를 활성화합니다. 실행 전에 `wandb login`을 통해 로그인해야 합니다.
사용을 원하지 않은 경우 false로 값을 넣으시면 됩니다

7. `--save_checkpoint=true` : 체크포인트를 저장하는 원하는 경우 사용하시면 됩니다.

8. `--save_freq=10000` : 몇 스텝 부터 저장 할지를 설정합니다.

```python
lerobot-train \
  --dataset.repo_id=HF_USER/TASK_NAME \
  --policy.repo_id=HF_USER/TASK_NAME \
  --policy.type=act \
  --policy.device=cuda \
  --job_name=act_so101 \
  --output_dir=outputs/train/act_so101/TASK_NAME \
  --wandb.enable=true \
  --batch_size=8 \
  --steps=100_000 \
  --save_checkpoint=true \
  --save_freq=10_000
```
